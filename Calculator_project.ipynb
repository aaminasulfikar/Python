```python
from tkinter import *
window=Tk()
window.title("Calculator")
window.configure(background='lightgray')
window.geometry('445x178')

expression=""

def press(num):
    global expression
    expression=expression+str(num)
    display_var.set(expression)
    
def clear():
    global expression
    expression=''
    display_var.set("")
    
def calculation_equals():
    try:
        global expression
        result=str(eval(expression))
        display_var.set(result)
        expression=" "
    except:
        display_var.set('Error')
expression=""

display_var=StringVar()
display=Entry(window,text=display_var,font=('Ariel bold',20),justify='right')
display.grid(row=0,column=0,columnspan=4,ipadx=70)

button1=Button(window,text='1',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press(1),width=7)
button1.grid(row=2,column=0)

button2=Button(window,text='2',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press(2),width=7)
button2.grid(row=2,column=1)

button3=Button(window,text='3',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press(3),height=1,width=7)
button3.grid(row=2,column=2)

plus=Button(window,text='+',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press('+'),height=1,width=7)
plus.grid(row=2,column=3)

button4=Button(window,text='4',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press(4),height=1,width=7)
button4.grid(row=3,column=0)

button5=Button(window,text='5',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press(5),height=1,width=7)
button5.grid(row=3,column=1)

button6=Button(window,text='6',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press(6),height=1,width=7)
button6.grid(row=3,column=2)

minus=Button(window,text='-',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press('-'),height=1,width=7)
minus.grid(row=3,column=3)

button7=Button(window,text='7',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press(7),height=1,width=7)
button7.grid(row=4,column=0)

button8=Button(window,text='8',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press(8),height=1,width=7)
button8.grid(row=4,column=1)

button9=Button(window,text='9',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press(9),height=1,width=7)
button9.grid(row=4,column=2)

multiply=Button(window,text='*',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press('*'),height=1,width=7)
multiply.grid(row=4,column=3)

bracket1=Button(window,text='(',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press('('),height=1,width=7)
bracket1.grid(row=5,column=0)

button0=Button(window,text='0',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press(0),height=1,width=7)
button0.grid(row=5,column=1)

bracket2=Button(window,text=')',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press(')'),height=1,width=7)
bracket2.grid(row=5,column=2)

div=Button(window,text='/',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press('/'),height=1,width=7)
div.grid(row=5,column=3)

clear_btn=Button(window,text='C',fg='black',bg='red',font=('Ariel bold',10),command=clear,height=1,width=7)
clear_btn.grid(row=6,column=1)

decimel=Button(window,text='.',fg='black',bg='red',font=('Ariel bold',10),command=lambda: press('.'),height=1,width=7)
decimel.grid(row=6,column=2)

equal=Button(window,text='=',fg='black',bg='red',font=('Ariel bold',10),command=calculation_equals,height=1,width=7)
equal.grid(row=6,column=3)
window.mainloop()  
    
```
